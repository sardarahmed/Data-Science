```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_openml
```

```python
df = fetch_openml('titanic', version=1, as_frame=True)['data']
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.set()
miss_value_percent= pd.DataFrame(df.isnull().sum()/len(df)*100) 
miss_value_percent.plot(kind='bar', title='Percentage of missing values by feature', ylabel='Percentage', xlabel='Features')
```

```python
print(f'shape of dateframe: {df.shape}')
```

```python
df.drop(['body'], axis = 1, inplace = True)
```

```python
print(f'shape of dateframe: {df.shape}')
```

```python
from sklearn.impute import SimpleImputer
```

```python
print(f'number of null values in age column before imputation: {df["age"].isnull().sum()}')
```

```python
imp = SimpleImputer(strategy= 'mean')
df['age'] = imp.fit_transform(df[['age']])
print(f'number of null values in age column after imputation: {df["age"].isnull().sum()}')
```
